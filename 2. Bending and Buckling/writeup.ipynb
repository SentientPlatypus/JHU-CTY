## Buckling

Buckling force is measured as $$P = \frac{E\pi^3r^4}{4L^2}$$

Where $E$ is Youngs modulus, $r$ is the raduius of the spaghetti, and $L$ is the length of the spaghetti

```python
buckling = pd.read_csv("buckling.csv")

regular = buckling[buckling["type"] == "REGULAR"]
thin = buckling[buckling["type"] == "THIN"]
angel = buckling[buckling["type"] == "ANGEL HAIR"]
buckling
```

We want to see the relationship between length of the spaghetti and the buckling load.

```python
line_colors = ['blue', 'green', 'red', 'orange', 'purple', 'brown']
labelnames = ["regular", "thin", "angel hair"]
labels = [regular, thin, angel]
for i in range(len(labels)):
    plt.plot(labels[i]["length (m)"], labels[i]["buckling load (N)"], color = line_colors[i], label = labelnames[i])

plt.title("Buckling load as a function of length")

plt.xlabel("Length of spaghetti")
plt.ylabel("Bucking load (N)")

plt.legend()

# Display the plot
plt.show()
```

We see that this relationship is similar to that of an inverse function is dilated by constant $k$, where $$k = \frac{E\pi^3r^4}{4}$$ <br>

rewriting the equation for buckling load, we get
$$P = \frac{k}{L^2} $$

We can linearlize this by taking the $\ln$ of $P$ and $L$ <br>
$$\ln P = \ln \left(\frac{k}{L^2}\right)$$
$$\ln P = \ln k - \ln \left(L^2\right)$$
$$\ln P = \ln k - 2\ln L \rightarrow E = \frac {4 e^{\ln k}}{\pi ^3 r^4}$$

```python
plt.xlim(-2, 2)
plt.ylim(-10, 10)

young_moduli_buckling = []

for i in range(len(labels)):
    x = labels[i]["length (m)"].apply(np.log)  # X-axis data
    y = labels[i]["buckling load (N)"].apply(np.log)  # Y-axis data

    # Perform linear regression (fit a polynomial line of degree 1)
    coeffs = np.polyfit(x, y, deg=1)

    # Extract the slope and intercept from the coefficients
    slope = coeffs[0]
    intercept = coeffs[1]

    # Create the equation string
    equation = f"ln yhat = {slope:.2f} * ln x + {intercept:.2f}"
    young_modulus = (math.exp(intercept) * 4) / (math.pi**3 * labels[i]["Radius (m)"].iloc[0]**4)
    young_moduli_buckling.append(young_modulus)
    # Print the equation for the label type
    print(f"Label Type: {labelnames[i]}")
    print("Radius: ", labels[i]["Radius (m)"].iloc[0])
    print("Equation:", equation)
    print(f"Young Modulus: {young_modulus:.2e}")
    print()


    rangevals = np.linspace(-2, 1, 50)
    # Plot the data points and the fitted line
    plt.scatter(x, y, color=line_colors[i], label=labelnames[i])
    plt.plot(rangevals, np.polyval(coeffs, rangevals), '--', color=line_colors[i])

# Set plot title and labels
plt.title("Buckling load as a function of length")
plt.xlabel("natural log of Length of spaghetti")
plt.ylabel("natural log Buckling load (N)")

# Display the legend
plt.legend()

# Display the plot
plt.show()
```

From Greatest to smallest young modulus:
1. Regular
2. Angel Hair
3. Thin

The Young Modulus constant is different for each of the different radiuses. This makes sense as a larger radius indicates a stronger material.

## Bending
Bending is represented as :
$$P = \frac{12\pi E r^4 y}{L^3} \rightarrow y = \frac{PL^3}{12\pi E r^4}$$
Where $P$ is bending load, $E$ is youngs modulus, $r$ is spaghetti radius, $y$ is deflection$ and $L$ is distance between supports

Like buckling, we can put this in terms of constant $k$ where
$$k = \frac{L^3}{12\pi E r^4} \rightarrow E = \frac {L^3} {k 12\pi r^4}$$ 

so,

$$y = kP$$

```python
bending = pd.read_csv("good_bending.csv")
print(bending.columns)
bending = bending.drop(['notes'], axis=1)
bending.dropna(inplace=True)
regular = bending[bending["type"] == "REGULAR"]
thin = bending[bending["type"] == "THIN"]
angel = bending[bending["type"] == "ANGEL HAIR"]
bending
```

Lets observe the relationship between load and deflection:

```python
line_colors = ['blue', 'green', 'red', 'orange', 'purple', 'brown']
labelnames = ["regular", "thin", "angel hair"]
young_moduli_bending = []
labels = [regular, thin, angel]
plt.ylim(-.1, .1)

for i in range(len(labels)):
    x = labels[i]["bending load (N)"] # X-axis data
    y = labels[i]["Deflection (m)"] # Y-axis data

    # Perform linear regression (fit a polynomial line of degree 1)
    coeffs = np.polyfit(x, y, deg=1)

    # Extract the slope and intercept from the coefficients
    slope = coeffs[0]
    intercept = coeffs[1]

    # Create the equation string
    equation = f"yhat = {slope:.2f} * x + {intercept:.2f}"
    young_modulus = (labels[i]["length (m)"].iloc[0] ** 3) / (slope * 12 * math.pi * labels[i]["Radius (m)"].iloc[0] ** 4)
    young_moduli_bending.append(young_modulus)
    

    # Print the equation for the label type
    print(f"Label Type: {labelnames[i]}")
    print("Radius: ", labels[i]["Radius (m)"].iloc[0])
    print("Distance in supports: ", labels[i]["length (m)"].iloc[0])
    print("Equation:", equation)
    print(f"Young Modulus: {young_modulus:.2e}")
    print()

    rangevals = np.linspace(0, .5, 10)
    # Plot the data points and the fitted line
    plt.scatter(x, y, color=line_colors[i], label=labelnames[i])
    plt.plot(rangevals, np.polyval(coeffs, rangevals), '--', color=line_colors[i])

# for i in range(len(labels)):
#     plt.plot(labels[i]["bending load (N)"], labels[i]["Deflection (m)"], color = line_colors[i], label = labelnames[i])

plt.title("Deflection as a function of load")

plt.xlabel("Bending load (N)")
plt.ylabel("Deflection (m)")

plt.legend()

# Display the plot
plt.show()
```

$$$$

From the Bending calculations, the Young Modulus we got in descending order is:
1. Regular
2. Thin
3. Angel hair

Unlike the young modulus values derived from buckling, these make a little more sense, given the fragility of the different spaghetti.

## Fracture-Strength

The radius of the arc formed by the bending is represented as:

$$\rho = \frac {\left(\frac{L}{2}\right)^2 + y^2}{2y}$$

Strain, $\epsilon$ can be represented as:
$$\epsilon = \frac{r}{\rho}$$

Where $C$ is the distance from the neutral plane.

to find maximum strain, $\epsilon_{max}$, we want the value of $\rho$ and $C$ when the spaghetti fractured.

Then, with maximum strain we can calculate maximum stress, $\sigma_{max}$,  with Hooke's law.
$$\sigma_{max} = E \epsilon_{max}$$

Finally, Fracture strength can be represented as 
$$P_{max} = \pi r^2 \sigma_{max}$$

```python
i = 0
fracture_strengths = []
radii = []
for type in [regular, thin, angel]:
    whenItBroke = type.loc[type["bending load (N)"].idxmax()]
    area = whenItBroke["Radius (m)"] ** 2 * math.pi
    arc = (((whenItBroke["length (m)"] / 2) **2) + (whenItBroke["Deflection (m)"]**2))/(2 * whenItBroke["Deflection (m)"])
    max_strain = whenItBroke["Radius (m)"] / arc
    max_stress = young_moduli_bending[i] * max_strain
    whenItBroke["fracture_strength"] = area * max_stress
    fracture_strengths.append(area * max_stress)
    radii.append(whenItBroke["Radius (m)"])
    print(whenItBroke["type"])
    print("fracture load: ", whenItBroke["bending load (N)"])
    print("fracture_strength: ", area * max_stress)
    print("radius: ", whenItBroke["Radius (m)"])
    print()
    plt.scatter(whenItBroke["Radius (m)"], whenItBroke["fracture_strength"], color = line_colors[i], label = labelnames[i])
    i += 1

plt.plot(radii, fracture_strengths)

plt.title("Fracture Strength vs Radius")
plt.xlabel("Radius (m)")
plt.ylabel("Fracture Strength")
plt.legend()
```

Clearly, fracture strength is different for each radius. As the radius increases, the fracture strength increases. This fits with the intuition that thicker materials are stronger. Additionally, the association is fairly linear and positive.

For this lab, we analyzed the buckling and bending for three different types of spaghetti. For Bending, We put each spaghetti between two tables, and added weight and measured displacement at each load. For buckling, we applied vertical force, and measured the length of the spaghetti and the force required for it to begin to bend.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math
```

## Buckling

Buckling force is measured as $$P = \frac{E\pi^3r^4}{4L^2}$$

Where $E$ is Youngs modulus, $r$ is the raduius of the spaghetti, and $L$ is the length of the spaghetti

```python
buckling = pd.read_csv("buckling.csv")

regular = buckling[buckling["type"] == "REGULAR"]
thin = buckling[buckling["type"] == "THIN"]
angel = buckling[buckling["type"] == "ANGEL HAIR"]
buckling
```

We want to see the relationship between length of the spaghetti and the buckling load.

```python
line_colors = ['blue', 'green', 'red', 'orange', 'purple', 'brown']
labelnames = ["regular", "thin", "angel hair"]
labels = [regular, thin, angel]
for i in range(len(labels)):
    plt.plot(labels[i]["length (m)"], labels[i]["buckling load (N)"], color = line_colors[i], label = labelnames[i])

plt.title("Buckling load as a function of length")

plt.xlabel("Length of spaghetti")
plt.ylabel("Bucking load (N)")

plt.legend()

# Display the plot
plt.show()
```

We see that this relationship is similar to that of an inverse function is dilated by constant $k$, where $$k = \frac{E\pi^3r^4}{4}$$ <br>

rewriting the equation for buckling load, we get
$$P = \frac{k}{L^2} $$

We can linearlize this by taking the $\ln$ of $P$ and $L$ <br>
$$\ln P = \ln \left(\frac{k}{L^2}\right)$$
$$\ln P = \ln k - \ln \left(L^2\right)$$
$$\ln P = \ln k - 2\ln L \rightarrow E = \frac {4 e^{\ln k}}{\pi ^3 r^4}$$

```python


for i in range(len(labels)):
    x = labels[i]["length (m)"].apply(np.log)  # X-axis data
    y = labels[i]["buckling load (N)"].apply(np.log)  # Y-axis data

    # Perform linear regression (fit a polynomial line of degree 1)
    coeffs = np.polyfit(x, y, deg=1)

    # Extract the slope and intercept from the coefficients
    slope = coeffs[0]
    intercept = coeffs[1]

    # Create the equation string
    equation = f"ln yhat = {slope:.2f} * ln x + {intercept:.2f}"

    young_modulus = (math.exp(intercept) * 4) / (math.pi**3 * labels[i]["Radius (m)"].iloc[0]**4)

    # Print the equation for the label type
    print(f"Label Type: {labelnames[i]}")
    print("Radius: ", labels[i]["Radius (m)"].iloc[0])
    print("Equation:", equation)
    print(f"Young Modulus: {young_modulus:.2e}")
    print()

    # Plot the data points and the fitted line
    plt.scatter(x, y, color=line_colors[i], label=labelnames[i])
    plt.plot(x, np.polyval(coeffs, x), '--', color=line_colors[i])

# Set plot title and labels
plt.title("Buckling load as a function of length")
plt.xlabel("natural log of Length of spaghetti")
plt.ylabel("natural log Buckling load (N)")

# Display the legend
plt.legend()

# Display the plot
plt.show()
```
